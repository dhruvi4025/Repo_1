```python
import pandas as pd
```

```python
data = {"Name": [" Alice ,Smith (CEO)"]} 
df = pd.DataFrame(data)
```

```python
start = df.loc[0, 'Name'].find("(")
end = df.loc[0, 'Name'].find(")")
df.loc[0, 'Name'][0:start] + df.loc[0, 'Name'][end+1:]
```

```python
#REPLACE SubStrings with Regex : regular expression 

# * means zero or more
# . match any character

"\(.*\)"

df['Name'].str.replace(r"\(.*\)","HELLO",regex=True).head()
```

```python
df['Name'].str.replace("\\(.*\\)", "HELLO", regex=True).head()
```

Regex pattern: r"\(.*\)"

\( → matches a literal opening parenthesis "(".

.* → matches any characters (zero or more).

\) → matches a literal closing parenthesis ")".

Together, \(.*\) means: “match everything inside parentheses, including the parentheses themselves.”

The r"" prefix makes it a raw string, so Python doesn’t misinterpret the backslashes.
